# Major data formats in the weather radar world


- CFRadial
- ODIM
- Rainbow
- Nexrad


# [CFRadial](https://www.eol.ucar.edu/system/files/CfRadialDoc-v2.0-20180430.pdf)

*Py-ART Radar object inherits CFRadial data structure*

- Based on NetCDF
- For **polar** radar coordinates only
- De-facto standard for the research community
- Standard format at WMO for radial weather data
- some CRFRadial readers: [Ï‰radlib](https://docs.wradlib.org/en/stable/generated/wradlib.io.backends.CfRadial2BackendEntrypoint.html), [Py-ART](https://arm-doe.github.io/pyart/API/generated/pyart.io.read_cfradial.html)

# [ODIM](https://www.eumetnet.eu/wp-content/uploads/2019/01/ODIM_H5_v23.pdf)

- OPERA (Europe weather radar cooperation) standard
- Based on HDF5
- Can be used for polar and Cartesian data
- some ODIM readers: [Ï‰radlib](https://docs.wradlib.org/en/stable/generated/wradlib.io.hdf.open_odim_mfdataset.html), [Py-ART](https://arm-doe.github.io/pyart/API/generated/pyart.aux_io.read_odim_h5.html) 



## ODIM data structure

<div>

</div>
<div style="float: left; font-size: 20pt">
 
- **How** : All instrument and acquisition parameters
- **Where** : e.g. WGS coordinates, range, azimuth (polar), grid size (Cartesian)
- **What** :    
   - at top-level: type of data (PVOL, CVOL,...), institution, file format
   - at dataset-level: product name, quantity, gain, offset
